Теперь осталось преобразовать валовую и бюджетную стоимость в реальные суммы в долларах с точки зрения покупательной способности 2016 года.

```python
from bs4 import BeautifulSoup
import requests

url = "https://www.usinflationcalculator.com/inflation/consumer-price-index-and-annual-percent-changes-from-1913-to-2008/"
r = requests.get(url)
data = r.text
soup = BeautifulSoup(data, "html.parser")

table = soup.find('table')
rows = table.tbody.findAll('tr')

years = []
cpis = []

for row in rows:
    year = row.findAll('td')[0].get_text()
    if year.isdigit() and int(year) < 2017:
        years.append(int(year))
        cpis.append(float(row.findAll('td')[13].get_text()))


cpi_table = pd.DataFrame({
    "year": years,
    "avg_annual_cpi": cpis
})

cpi_table.tail()
```

```python
def get_real_value(nominal_amt, old_cpi, new_cpi):
    real_value = (nominal_amt * new_cpi) / old_cpi
    return real_value


CPI = (cpi_table[cpi_table['year'] == 2016]['avg_annual_cpi'])
for x in CPI: CPI_2016 = x
# print(CPI_2016)

real_domestic_gross = []
real_budget_values = []

# должны преобразовать валовые и бюджетные значения 
# в реальное долларовое выражение 2016 г.

for index,row in df.iterrows():
    gross = row['gross']
    budget = row['budget']
    year = row['title_year']
    cp = (cpi_table[cpi_table['year'] == int(year)]['avg_annual_cpi'])

    for c in cp: cpi = c #равно единственному элементу Series cp

    real_gross = get_real_value(gross, cpi, CPI_2016)
    real_budget = get_real_value(budget, cpi, CPI_2016)
    real_domestic_gross.append(real_gross)
    real_budget_values.append(real_budget)
# print(real_domestic_gross)
df["real_domestic_gross"] = real_domestic_gross
df["real_budget"] = real_budget_values

profits = []
roi_vals = []

for index, row in df.iterrows():
    if row["real_domestic_gross"] !=0 and row["real_budget"] != 0:
        profit = row["real_domestic_gross"] - row["real_budget"]
        budget = row['real_budget']
        # конвертация ROI в проценты
        roi = (profit / budget) * 100
    else: profit = roi = 0

    profits.append(profit)
    roi_vals.append(roi)

df['profit'] = profits
df['roi'] = roi_vals

df.head(100)
```

# **Этап 3**. **Проверка гипотез**

# *Гипотеза 1: Окупшившеся фильмы имеют рейтинг IMDB больше 6*


Окупились - значит, значений поля profit - положительно

```python
imdb_scores = [] # здесь будут храниться рейтинги всех окупившихся фильмов

for i in range(len(df)):  #remove the thing
    if df['profit'][i] > 0: imdb_scores.append(df['imdb_score'][i])
```

Давайте узнаем, сколько оценок, меньших или равных 6, в процентном соотношении.

```python
wrong_counter = 0 #счётчик оценок, меньших или равных 6
correct_counter = 0 #счётчик оценок, больших 6
unknown_counter = 0 #счётчик неизвестных оценок (пропущенные значения)

for i in range(len(imdb_scores)):  #!!Сделать без цикла
    if imdb_scores[i] > 6: correct_counter += 1
    elif imdb_scores[i] == 0: unknown_counter += 1
    else: wrong_counter += 1
print(f'кол-во оценок, меньших или равных 6: {wrong_counter}')
print(f'кол-во оценок, больших 6: {correct_counter}')
print(f'кол-во оценок неизвестных оценок: {unknown_counter}')
```

Теперь вычислим процент всех неверных оценок, не согласованных с гипотезой:

```python
print(wrong_counter / correct_counter *100) #edit
```

Таким образом, приблизительно 33% от всех окупаемых фильмов имеют рейтинг IMDB, меньший или равный 6. Получается, первая гипотеза правдива примерно на 100-33=67%.

# *Гипотеза 2: Более 50% всех фильмов, выпущенных США, окупились*

Более 50% - значит для того, чтобы гипотеза была верна, необходимо, что окупившихся фильмов США было больше половины от всех фильмов, выпущенных США

```python
positive_profit_counter = 0
negative_profit_counter = 0
unknown_profit_counter = 0
for i in range(len(df)):
    if df['country'][i] == "USA": 
        if df['profit'][i] > 0:
            positive_profit_counter += 1
        elif df['profit'][i] == 0: unknown_profit_counter += 1
        else: negative_profit_counter += 1
            
print(f'кол-во окупившихся в США: {positive_profit_counter}')
print(f'кол-во неокупившихся в США: {negative_profit_counter}')
print(f'кол-во неизвестных полей profit в США: {unknown_profit_counter}')
print(f'общее кол-во фильмов США: {negative_profit_counter + positive_profit_counter + unknown_profit_counter}')

print(f'процент окупившихся от всех фильмов США без учёта неизвестных полей: {round(positive_profit_counter / (negative_profit_counter + positive_profit_counter) *100,2)}%')
print(f'процент окупившихся от всех фильмов США с учётом неизвестных полей: {round(positive_profit_counter / (negative_profit_counter + positive_profit_counter+unknown_profit_counter) *100,2)}%')
 
```

Как видим из последнего вывода:
- Без учета фильмов с неизвестным полем profit окупившихся фильмов больше 50%  
- C учётом фильмов с неизвестным полем profit окупившихся фильмов меньше 50% от всех фильмов- 46.23%. 
- Кол-во фильмов США с неизвестным полем profit достаточно, чтобы либо неокупившихся фильмов стало больше, либо наоборот - окупившихся. 

Можно, например, посчитать, сколько необходимо добавить из неизвестных в положительный профит, чтобы окупившихся фильмов стало больше половины:
Давайте высчитаем какого количества неизвестных полей профита не хватает положительным - окупаемым фильмам, чтобы процент был больше 50:

```python
counter = 0 # счётчик кол-ва, которое необходимо добавить к положительным профитам от неизвестных
current_percentage = round(positive_profit_counter / (negative_profit_counter + positive_profit_counter+unknown_profit_counter) *100,2)
NEW_positive_profit_counter = positive_profit_counter # положительное кол-во профитов, которое будем инкрементировать

#убрать цикл
while current_percentage <= 50: # пишем условие, при котором мы в цикле инкрементируем кол-во полож профитов 
    NEW_positive_profit_counter += 1 # пока процент не превысит 50
    counter += 1
    current_percentage = round(NEW_positive_profit_counter / (negative_profit_counter + positive_profit_counter + unknown_profit_counter) *100,2)

print(f'кол-во которое необходимо добавить к положительным профитам от неизвестных, чтобы процент положительных профитов превысил 50: {counter}')
print(current_percentage)
```

Таким образом, можно сделать вывод, что из 714 фильмов США с неизвестным полем "profit" достаточно только чтобы 141 фильм с неизвестным полем "profit" перешёл в положительный профит - окупился, чтобы окупившихся фильмов США было больше половины.  

Итак, подводя итог:
- без учёта фильмов с неизвестными полями - гипотеза верна
- с учётом фильмов с неизвестными полями нельзя точно сказать - гипотеза и не верна и не неверна.

Таким образом, гипотеза верна частично.
